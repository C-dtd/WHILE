```python
from flask import *
from flask_cors import CORS
from PIL import Image
import json
import os
```

```python
app = Flask(__name__)
CORS(app, resources={r'*': {'origins': 'localhost:5050'}})

@app.route('/imgupload', methods = ['get', 'post'])
def upload():
    with open('./imgid.json', 'r', encoding = 'UTF-8') as f:
        imgid = json.load(f)

    name = imgid['LASTID']
    imgid['LASTID'] += 1
    
    img = Image.open(request.files['img'])
    img.save(f'static/img/{name}.webp', format = 'webp', quality = 90)

    with open('./imgid.json', 'w', encoding = 'UTF-8') as f:
        json.dump(imgid, f, indent = '\t', ensure_ascii = False)
    return jsonify({"img_id": str(name)})

@app.route('/getimg', methods = ['get', 'post'])
def getimg():
    imgid = request.args.get('id', '0')
    if os.path.exists(f'./static/img/{imgid}.webp'):
        return send_file(f'./static/img/{imgid}.webp', mimetype='image/webp')
    else:
        return send_file('./static/img/0.webp', mimetype='image/webp')

@app.route('/')
def index():
    return render_template('index.html')
app.run(host='localhost', port=5051)
```
